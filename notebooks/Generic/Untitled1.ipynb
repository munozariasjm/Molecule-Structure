```python
nt = 1001
x = np.linspace(-1,1,nt)
sig = 0.2
signal = np.exp(-x**2/(2*sig**2))
mean = 0
std = 0.3
scale = 1
g_noise = scale*np.random.normal(mean, std, size=nt)
w_noise = scale*(np.random.random(size=nt)-0.5)
```

```python
plt.figure()
plt.plot(x,signal)

plt.figure()
plt.plot(x,signal+g_noise)

plt.figure()
plt.plot(x,signal+w_noise)
```

```python
n_runs = 10
wn_array = []
gn_array = []
for i in range(n_runs):
    wn_array.append(signal+scale*(np.random.random(size=nt)-0.5))
    gn_array.append(signal+scale*np.random.normal(mean, std, size=nt))
wn_array = np.array(wn_array)
gn_array = np.array(gn_array)
```

```python
#integrate first
wn_s = np.sum(wn_array,axis=1)/nt
gn_s = np.sum(gn_array,axis=1)/nt
print('white noise mean,std:', wn_s.mean(),wn_s.std())
print('gaussian noise mean,std:', gn_s.mean(),gn_s.std())
```

```python
#avg first
#integrate first
wn_m = np.mean(wn_array,axis=0)
gn_m = np.mean(gn_array,axis=0)
wn_std = np.std(wn_array,axis=0)
gn_std = np.std(gn_array,axis=0)
print('white noise mean,std:', (wn_m.sum()/nt),(np.sqrt((wn_std**2).sum())/nt))
print('gaussian noise mean,std:', (gn_m.sum()/nt),(np.sqrt((gn_std**2).sum())/nt))
```
